```python
def subplot_graph(df, features):
    f_count = len(features)
    cols = f_count
    if f_count < 4:
        rows = 1
    else:
        rows = math.ceil(f_count/3)
        cols = 3
                
    # Set up the matplotlib figure
    fig, axes = plt.subplots(rows, cols, figsize=(16, 4*rows))
        
    sub_rows = 0
    sub_cols = 0    
    for f in features:
        if sub_cols > cols-1:
            sub_cols = 0
            sub_rows += 1
            
        unique_vals = len(np.unique(df[f]))
        if unique_vals < 10:
            df_sub = df.groupby(df.columns[-1])[f].value_counts().sort_index(level=0)
            x_no = list(df_sub.xs(0, level=0).index)
            y_no = list(df_sub.xs(0, level=0).values)
            x_yes = list(df_sub.xs(1, level=0).index)
            y_yes = list(df_sub.xs(1, level=0).values)
            list_no = list(zip(x_no, y_no))
            list_yes = list(zip(x_yes, y_yes))
            df_no = pd.DataFrame(list_no)
            df_yes = pd.DataFrame(list_yes)

            df_plot = pd.merge(df_no, df_yes, on=df_no.columns[0], how="outer")
            df_plot = df_plot.fillna(0)
            try:
                axes[sub_cols].bar(df_plot.iloc[:,0], df_plot.iloc[:,1], label='0', color='firebrick')
                axes[sub_cols].bar(df_plot.iloc[:,0], df_plot.iloc[:,2], bottom=df_plot.iloc[:,1], 
                                   label='1', color='steelblue')
                axes[sub_cols].legend(title=df.columns[-1])
                axes[sub_cols].set_title(f)
                axes[sub_cols].set_xticks(df[f].unique().tolist(), minor=False)
            except:
                axes[sub_rows, sub_cols].bar(df_plot.iloc[:,0], df_plot.iloc[:,1], label='0', color='firebrick')
                axes[sub_rows, sub_cols].bar(df_plot.iloc[:,0], df_plot.iloc[:,2], bottom=df_plot.iloc[:,1],label='1', color='steelblue')  
                axes[sub_rows, sub_cols].legend(title=df.columns[-1])
                axes[sub_rows, sub_cols].set_title(f)
                axes[sub_rows, sub_cols].set_xticks(df[f].unique().tolist(), minor=False)
        else:    
            df_3 = df[df[df.columns[-1]] == 0][f]
            df_4 = df[df[df.columns[-1]] == 1][f]
            try:
                df_3.plot(ax=axes[sub_cols], kind='kde', color='red', label='0')
                df_4.plot(ax=axes[sub_cols], kind='kde', color='steelblue', label='1')
                axes[sub_cols].hist(df_3, bins=20, density=True, stacked=True, color='lightgreen')
                axes[sub_cols].set_title(f)
                axes[sub_cols].legend(title=df.columns[-1])
                axes[sub_cols].set_xlim(xmin=0)
            except:
                df_3.plot(ax=axes[sub_rows, sub_cols], kind='kde', color='red', label='0')
                df_4.plot(ax=axes[sub_rows, sub_cols], kind='kde', color='steelblue', label='1')
                axes[sub_rows, sub_cols].hist(df_3, bins=20, density=True, stacked=True, color='lightgreen')
                axes[sub_rows, sub_cols].set_title(f)
                axes[sub_rows, sub_cols].legend(title=df.columns[-1])
                axes[sub_rows, sub_cols].set_xlim(xmin=-5, xmax=100)
            
        sub_cols += 1
        
    fig.tight_layout()
    fig.show()

features = ['Season', 'Age', 'Childish diseases', 'Accident or serious trauma', 'Surgical intervention',
            'High fevers in the last year','Frequency of alcohol consumption','Smoking habit',
            'Number of hours spent sitting per day']
subplot_graph(df, features)
```

```python
def create_heatmap(hm, figsize=(20, 19)):
    fig, ax = plt.subplots(figsize=figsize)

    im = ax.imshow(hm, 
    #                vmin=0, vmax=10, 
                   cmap='viridis', aspect='auto')

    # Create colorbar
    cbar = ax.figure.colorbar(im, ax=ax)

    # We want to show all ticks...
    ax.set_xticks(np.arange(len(hm.columns)))
    ax.set_yticks(np.arange(len(hm.columns)))
    # ... and label them with the respective list entries
    ax.set_xticklabels(hm.columns)
    ax.set_yticklabels(hm.columns)

    # Turn spines off and create white grid.
    ax.spines[:].set_visible(False)
    ax.set_xticks(np.arange(hm.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(hm.shape[0]+1)-.5, minor=True)
    ax.grid(which="minor", color="w", linestyle='-', linewidth=3)
    ax.tick_params(which="minor", bottom=False, left=False)

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    for i in range(len(hm.columns)):
        for j in range(len(hm.columns)):
            hm_val = round(hm.values[i, j], 2)
            if hm_val > 0.85:
                text = ax.text(j, i, hm_val,
                               ha="center", va="center", color="black", size=16)
            else:
                text = ax.text(j, i, hm_val,
                               ha="center", va="center", color="w", size=16)

    fig.tight_layout()
    plt.show()
```

```python
# Confusion Matrix

def confusion_matrix_func(cm, cm_title):
    fig, ax = plt.subplots(figsize=(4, 4))

    # Hiển thị heatmap
    im = ax.imshow(cm, interpolation='nearest', cmap='Reds', aspect='auto')

    # Hiển thị các dấu ticks
    ax.set_xticks(np.arange(len(cm.tolist())))
    ax.set_yticks(np.arange(len(cm.tolist())))


    thresh = cm.max() / 1.5
    # Lặp lại kích thước dữ liệu và tạo chú thích văn bản
    for i in range(len(cm.tolist())):
        for j in range(len(cm.tolist())):
            text = ax.text(j, i, cm.tolist()[i][j],
                           ha="center", va="center", size=16,
                           color="white" if cm[i, j] > thresh else "black")

    # Để nhãn xuất hiện trên cùng
    ax.xaxis.set_ticks_position('top')
    ax.xaxis.set_label_position('top')

    plt.xlabel('Actual value', size=16)
    plt.ylabel('Predicted value', size=16)
    plt.title(cm_title, size=20, x=0.2, y=1.2)
    plt.show()
```

```python
def Confusion_matrix_metrics(TP, FP, FN, TN):
    # Sensitivity(Độ nhạy) , hit rate(tỷ lệ trúng) , recall(thu hồi) , or true positive rate(tỷ lệ tích cực thực sự)
    TPR = TP / (TP + FN)
    print('The True Positive Rate is: {:.2%}'.format(TPR))
    # Specificity(Độ đặc hiệu) , selectivity(Độ chọn lọc) or true negative rate (TNR)(tỷ lệ tiêu cực thực sự)
    TNR = TN / (TN + FP)
    print('The True Negative Rate is: {:.2%}'.format(TNR))
    print('='*10)

    # accuracy (ACC) (Chính xác)
    ACC = (TP + TN) / (TP + TN + FP + FN)
    print('The Accuracy is: {:.2%}'.format(ACC))
    # balanced accuracy (BA) (Chính xác cân bằng)
    BA = (TPR + TNR) / 2
    print('The Balanced Accuracy is: {:.2%}'.format(BA))
    print('='*10)

    # Precision or positive predictive value (Giá trị tiên đoán chính xác hoặc tích cực)
    PPV = TP / (TP + FP)
    print('The Precision is: {:.2%}'.format(PPV))
    # negative predictive value (NPV) (Giá trị tiên đoán sai)
    NPV = TN / (TN + FN)
    print('The Negative Predictive Value is: {:.2%}'.format(NPV))
    # false discovery rate (FDR) (Tỷ lệ phát hiện sai)
    FDR = 1 - PPV
    print('The False Discovery Rate is: {:.2%}'.format(FDR))
    # false omission rate (FOR) (Tỷ lệ bỏ sót sai)
    FOR = 1 - NPV
    print('The False Omission Rate is: {:.2%}'.format(FOR))
    print('='*10)

    # prevalence threshold (PT) (Ngưỡng phổ biến)
    PT = (math.sqrt(TPR*(1 - TNR)) + TNR - 1)/(TPR + TNR - 1)
    print('The Prevalence Threshold is: {:.2}'.format(PT))
    # F1 score
    F1 = 2*TP / (2*TP + FP + FN)
    print('The F1 Score is: {:.2}'.format(F1))
    # Matthews correlation coefficient (MCC) or phi coefficient (Hệ số tương quan)
    MCC = ((TP*TN) - (FP*FN)) / math.sqrt((TP + FP)*(TP + FN)*(TN + FP)*(TN + FN))
    print('The Matthews Correlation Coefficient is: {:.2}'.format(MCC))
    print('='*10)

    # False positive rate or False alarm rate (Tỷ lệ dương và cảnh báo sai)
    FPR = FP / (FP + TN)
    print('The False positive rate is: {:.2}'.format(FPR))
    # False negative rate or Miss Rate (Tỷ lệ âm và trượt)
    FNR = FN / (FN + TP)
    print('The False Negative Rate is: {:.2%}'.format(FNR))
```
